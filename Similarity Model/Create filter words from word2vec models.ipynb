```python
from gensim.models import Word2Vec
import glob
from config import *
```

```python
models = glob.glob(TEMP_PATH + '/all_reuters_news_articles_*.w2v')
```

```python
models
```

```python
vol_words  = dict()
infl_words = dict()
gdp_words  = dict()

for model in models:
    year = model[-8:-4]
    model = Word2Vec.load(model)
    vol_words[year] = [word for (word, similarity) in model.wv.most_similar('volatility')]
    infl_words[year] = [word for (word, similarity) in model.wv.most_similar('inflation')]
    gdp_words[year] = [word for (word, similarity) in model.wv.most_similar('gdp')]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(vol_words).T.sort_index().T
df.to_csv(TEMP_PATH + '/volatility_words.csv')
```

```python
df = pd.DataFrame(infl_words).T.sort_index().T
df.to_csv(TEMP_PATH + '/inflation_words.csv')
```

```python
df = pd.DataFrame(gdp_words).T.sort_index().T
df.to_csv(TEMP_PATH + '/gdp_words.csv')
```
